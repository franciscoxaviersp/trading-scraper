```python
from bs4 import BeautifulSoup
import yfinance as yf
import requests

week_start = "2023-11-26"
week_end = "2023-12-02"
day = "2023-11-29"

base_url = "https://finance.yahoo.com/calendar/earnings?from={week_start}&to={week_end}&day={day}"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:120.0) Gecko/20100101 Firefox/120.0"}

url = base_url.format(week_start=week_start, week_end=week_end, day=day)

response = requests.get(url, headers=headers)


soup = BeautifulSoup(response.content, "html.parser")

results = soup.find(id="cal-res-table")
#print(results.prettify())

job_elements = results.find_all("tr", class_="simpTblRow")

#print(job_elements)

data = {}

for element in job_elements:
    # get all td elements from element
    td_elements = element.find_all("td")
    print(td_elements)
    # get company name from 2nd td element
    
    symbol = td_elements[0].text
    company_name = td_elements[1].text
    event_name = td_elements[2].text
    earnings_call_time = td_elements[3].text
    eps_estimate = td_elements[4].text
    reported_eps = td_elements[5].text
    surprise = td_elements[6].text
    
    yf_info = yf.Ticker(symbol)
    
    print(yf_info.earnings_dates)
    
    data[symbol] = {
        "company_name": company_name,
        "event_name": event_name,
        "earnings_call_time": earnings_call_time,
        "eps_estimate": eps_estimate,
        "reported_eps": reported_eps,
        "surprise": surprise
    }
    
    break
    

```
